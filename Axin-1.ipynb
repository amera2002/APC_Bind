```python
def add_copy_numbers(row):

    prot_info.loc[prot_info['Tax ID']==row['Tax ID'],'Copy number']=row['copies']
```

```python
prot_info=pd.read_csv('/cluster/dtu/ug/aelnakla/cluster_data/axin/Axin-1.filtered-2.tsv', sep="\t")
prot_info['copies']=1
prot_info.apply(add_copy_numbers, axis=1)
prot_info
```

```python
accessions=list(prot_info['Accession'])
```

```python
records=SeqIO.parse('../cluster_data/axin/Axin-1.fasta', format='fasta')
output_records=list()
for record in records:
    rec_id=record.id.split('|')[0]
    if rec_id in accessions:
        record.id=rec_id
        record.description=''
        output_records.append(record)
```

```python
with open('../cluster_data/axin/Axin-1.filtered.fasta', 'w') as out_handle:
    SeqIO.write(output_records, out_handle, 'fasta')
```
